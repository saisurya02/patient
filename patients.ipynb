```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
df = pd.read_csv("patients.csv")
```

```python
print("Data Summary:")
print(df.describe(include='all'))
print("\n")
```

```python
satisfaction = df["satisfaction"]




mean_val = satisfaction.mean()
variance_val = satisfaction.var()
skewness_val = skew(satisfaction)
kurtosis_val = kurtosis(satisfaction)




print("=== Statistical Moments (Satisfaction) ===")
print(f"Mean: {mean_val:.2f}")
print(f"Variance: {variance_val:.2f}")
print(f"Skewness: {skewness_val:.2f}")
print(f"Kurtosis: {kurtosis_val:.2f}\n")
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="age", y="satisfaction", hue="service")
plt.title("Relational Plot: Age vs Satisfaction by Service")
plt.xlabel("Age")
plt.ylabel("Satisfaction Score")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="service", y="satisfaction", estimator="mean",)
plt.title("Categorical Plot: Average Satisfaction by Service")
plt.xlabel("Hospital Service")
plt.ylabel("Average Satisfaction")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
corr = df[["age", "satisfaction"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Statistical Plot: Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="service", y="satisfaction")
plt.title("Boxplot: Satisfaction by Service")
plt.xticks(rotation=45)
plt.show()
```
